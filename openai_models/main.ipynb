### Using better prompts for website generation###

```python
from openai import OpenAI
import os
import time
```

```python
api_key = os.getenv("OPENAI_API_KEY")
client = OpenAI(api_key=api_key)
```

```python
from website_gen_prompt import PROMPT
import asyncio

models = ["gpt-5.2", "gpt-5-mini", "gpt-5-nano", "gpt-5.2-codex", "gpt-5.1-codex-max"]
time_str = ''

for model in models:
    print(f"Generating website with {model} model...")
    time_start = time.time()

    # Assume that client.responses.create is async-compatible
    response = client.responses.create(
        model=model,
        input=PROMPT
    )

    time_end = time.time()
    time_diff = time_end - time_start
    print(f"Time taken ({model}): {time_diff} seconds")

    response_text = response.output_text

    file_path = f"improved_websites/website_{model}.html"
    with open(file_path, "w") as f:
        f.write(response_text)

    time_str += f"{model}: {time_diff} seconds\n"

print(time_str)
with open("improved_websites/time_str.txt", "w") as f:
    f.write(time_str)
    
```

### Deep Research ###

```python
from openai import OpenAI
from time import sleep
from deep_research_prompt import PROMPT
```

```python
client = OpenAI(timeout=3600)

resp = client.responses.create(
  model="gpt-5.2",
  input=PROMPT,
  background=True,
)

while resp.status in {"queued", "in_progress"}:
  print(f"Current status: {resp.status}")
  sleep(2)
  resp = client.responses.retrieve(resp.id)

print(f"Final status: {resp.status}\nOutput:\n{resp.output_text}")
```

**Deep Research Takeaways:**
- took 1:45 m to run.
- provided okay results, but this is not deep research even though I provided a list of questions.
- included references, but some of them retuned 404.
- all in all the result is not as suffisticated as I expected. I guess that maybe the subject should be narrower to get more into detail. 
- retuned 5k tokens, which is only 5% of its 100k output window. 

### Audio and Speech models ###

```python
import base64
from openai import OpenAI

client = OpenAI()

completion = client.chat.completions.create(
    model="gpt-audio",
    modalities=["text", "audio"],
    audio={"voice": "marin", "format": "wav"},
    messages=[
        {
            "role": "user",
            "content": "Provide an example of a small talk in an office kitchen between two people. Limit to 15 seconds. Do not provide any additional information or personal comments. Use slighly different voice or tone for each person. Do not announce the start of the conversation or who is speaking. Make one person raging and speaking quicklyand another super calm speaking smoothly and slowly."
        }
    ]
)

print(completion.choices[0])

wav_bytes = base64.b64decode(completion.choices[0].message.audio.data)
with open("conversation.wav", "wb") as f:
    f.write(wav_bytes)
```

Usage of realtime speech-to-speech agents required OpenAI Agent Kit - SDK for developing agents. Also, example of realtime agent is only available in Typescript. 

```python
from openai import OpenAI

client = OpenAI()
audio_file= open("conversation.wav", "rb")

transcription = client.audio.transcriptions.create(
    model="gpt-4o-transcribe", 
    file=audio_file
)

print(transcription.text)
```

```python
import base64
from openai import OpenAI

client = OpenAI()

def to_data_url(path: str) -> str:
    with open(path, "rb") as fh:
        return "data:audio/wav;base64," + base64.b64encode(fh.read()).decode("utf-8")

with open("conversation.wav", "rb") as audio_file:
    transcript = client.audio.transcriptions.create(
        model="gpt-4o-transcribe-diarize",
        file=audio_file,
        response_format="diarized_json",
        chunking_strategy="auto",
    )

for segment in transcript.segments:
    print(segment)
```

gpt-4o-transcribe-diarize failed to distinquish between two similar (almost same voices) that simulated a conversation. This was expected. However, after I asked the gpt-audio model to make two people speak wth different intonations, even though the voice is the same, transcription model managed to distinguish between two people. 

```python
import base64
from openai import OpenAI

client = OpenAI()

def to_data_url(path: str) -> str:
    with open(path, "rb") as fh:
        return "data:audio/wav;base64," + base64.b64encode(fh.read()).decode("utf-8")

with open("converation_with_different_voices.mp3", "rb") as audio_file:
    transcript = client.audio.transcriptions.create(
        model="gpt-4o-transcribe-diarize",
        file=audio_file,
        response_format="diarized_json",
        chunking_strategy="auto",
    )

for segment in transcript.segments:
    print(segment)
```

When voices are clearly different, model did great. It coul also assign names if I provided examples of speech of each converstion member. It could alos translate the audio or provide timestamps.
